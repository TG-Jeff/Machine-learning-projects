```python
#you can then split the data into features and labels
data = sonar.drop(60,axis=1)
label = sonar[60]

```

```python
#perform preprocessing such as converting the labels to int
#split the data into test and train
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
encoder = LabelEncoder()
label = encoder.fit_transform(label)

x_train,x_test,y_train,y_test = train_test_split(data,label,random_state=1,test_size=0.35)


```

```python
#we are gooing to use the random forest classifier as our model
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV



model = RandomForestClassifier() 
model.fit(x_train,y_train)
pred = model.predict(x_test)
print(metrics.accuracy_score(y_test,pred))
```

```python
#use the gridsearch to find appropriate parameters to improve model accuracy
forest_param = {
    'max_depth': list(np.arange(10,30)),
    'max_features': [10,15,20,25,30,35,40],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'random_state':[0,1,42]
}

grid = GridSearchCV(model,forest_param,cv=10,scoring='accuracy')
grid.fit(x_train,y_train)

print(grid.best_params_)
print(grid.best_score_)
```
